```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import ast
```

```python
# Load the data
data = pd.read_csv("train.csv", low_memory=False)
data.drop_duplicates() # remove duplicates

location = data.iloc[: , 21:34]
# drop Location.Address.UnparsedAddress
location = location.drop(columns=['Location.Address.UnparsedAddress'])
location
```

```python
def cleanColumn(column):
    # Eliminar les categories que apareixen menys d'un 5% de les vegades
    value_counts = column.value_counts()
    threshold = value_counts.max() * 0.05
    to_replace = value_counts[value_counts < threshold].index
    return column.replace(to_replace, 'Others')

```

```python
# concat locations with price
location_n_price = data['Listing.Price.ClosePrice'] # preu dels train

# concat prices with locations
location_n_price = pd.concat([location, location_n_price], axis=1)

location_n_price
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plott(location_n_price, column):
    
    # histoplot of the column with the price
    sns.histplot(data=location_n_price, x=column, y='Listing.Price.ClosePrice')
    plt.show()
    # scatter plot
    sns.scatterplot(data=location_n_price, x=column, y='Listing.Price.ClosePrice')
    plt.show()
```

```python

    

# Itera sobre cada columna numÃ¨rica per fer un scatter plot
for col in location_n_price:
    if col != 'Listing.Price.ClosePrice':  # Assumeix que la columna 'precio' Ã©s la target
        c = cleanColumn(data[col])
        plott(location_n_price, c)
    # AttributeError: 'StrCategoryFormatter' object has no attribute 'set_scientific'
    # https://stackoverflow.com/questions/6774086/why-is-my-xlabel-cut-off-in-my-matplotlib-plot
    
```

```python
# Convert non-numeric columns to numeric or drop them
location_n_price_numeric = location_n_price.apply(pd.to_numeric, errors='coerce')

# print corr with price
corr = location_n_price_numeric.corr()
print(corr['Listing.Price.ClosePrice'])
```
