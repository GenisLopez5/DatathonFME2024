```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
```

```python
# Llegeix el CSV
csv_file = 'train.csv'  # Canvia això per la ruta al teu fitxer CSV
df = pd.read_csv(csv_file)
```

```python
# Configura els gràfics
plt.style.use('seaborn')  # Estil per als gràfics
sns.set_palette("muted")  # Paleta de colors
```

```python
# Identifica columnes numèriques i categòriques
columnes_numeriques = df.select_dtypes(include=['number']).columns
columnes_categoriques = df.select_dtypes(exclude=['number']).columns
```

```python
# Itera sobre cada columna numèrica per fer un scatter plot
for col in columnes_numeriques:
    if col != 'Listing.Price.ClosePrice':  # Assumeix que la columna 'precio' és la target
        plt.figure(figsize=(8, 6))
        sns.scatterplot(data=df.dropna(subset=col), x=col, y='Listing.Price.ClosePrice')

        # Configura el format dels ticks de l'eix Y (preu)
       # plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))  # Separadors de milers
        plt.ticklabel_format(style='plain', axis='y')  # Evita notació científica a l'eix Y
        plt.ticklabel_format(style='plain', axis='x')  # Evita notació científica a l'eix X si cal

        plt.title(f'Relació entre {col} i Preu')
        plt.xlabel(col)
        plt.ylabel('Preu')
        plt.show()
```
